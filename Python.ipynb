```python
legs=int(input("Enter no of legs"))
heads=int(input("Enter no of heads"))
flag=False
for cows in range(0,heads+1):
  hens=heads-cows
  cal_legs=cows*4 + hens*2
  if cal_legs==legs:
    flag=True
    break
if flag:
  print("Cows:",cows)
  print("Hens:",hens)
else:
  print("No solution")
```

```python
#DIGITAL TIME
import time
import sys
h=3
s=0
m=40
while True:
  sys.stdout.write("\r{:02d}:{:02d}:{:02d}".format(h,m,s))
  sys.stdout.flush()
  time.sleep(1)
  s=s+1
  if s==60:
    s=0
    m=m+1
  if m==60:
    m=0
    h=h+1
  if h==24:
    h=0
```

```python
names=["Ram","Raj","Rose","Jack","Pavi"]
marks=[[45,33,90],[27,78,38],[87,56,47],[96,57,62],[92,64,23]]
for i in range(5):
  p=sum(marks[i])//3
  if p>80:
    g="S"
  elif p<40:
    g="F"
  elif p>40 and p<60:
    g="B"
  else:
    g="A"
  print("{}.Mr/Ms {} has scored {}%-{} Grade".format(i+1,names[i].title(), p, g))
```

```python
import random
a=int(input("Enter num of teams: "))
teams=[]
print("Enter all teams name: ")
for i in range(a):
  t=input("Enter team name: ")
  teams.append(t)
m=int(input("Enter num of meet bw teams: "))
matches = []
for i in range(0,a-1):
  for j in range(i+1,a):
    for k in range(m):
      matches.append([teams[i],teams[j]])
      random.shuffle(matches)
      pos=1
  for match in matches:
    print("MATCH {}:{} vs {}".format(pos,match[0],match[1]))
  pos=pos+1
```
