```python
legs=int(input("Enter no of legs"))
heads=int(input("Enter no of heads"))
flag=False
for cows in range(0,heads+1):
  hens=heads-cows
  cal_legs=cows*4 + hens*2
  if cal_legs==legs:
    flag=True
    break
if flag:
  print("Cows:",cows)
  print("Hens:",hens)
else:
  print("No solution")
```

```python
#Dependent Loop
for i in range(100,1000):
  is_prime=True
  for j in range(2, i):
    if i % j == 0:
      is_prime=False
      break
  if is_prime:
    print(i)
```
